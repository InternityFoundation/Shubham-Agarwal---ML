# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.svm import SVR, LinearSVR
```

# Loading Dataset

```python
data = load_boston()
```

```python
data.feature_names
```

```python
X = data.data
y = data.target
```

```python
print(X.shape)
print(y.shape)
```

# Visualizing Dataset

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']

plt.subplot(1, 2 , 1)
plt.scatter(X[:, 5], y, marker='o')
plt.title('RM')
plt.xlabel('RM')
plt.ylabel('Y')

plt.subplot(1, 2 , 2)
plt.scatter(X[:, -1], y, marker='o')
plt.title('LSTAT')
plt.xlabel('LSTAT')
plt.ylabel('Y')
```

# Training Linear Model

```python
svr = LinearSVR(epsilon=2)
svr.fit(X, y)
```

# Training Non Linear Model

```python
regressor = SVR(kernel='rbf', C=100, epsilon=1.5)
regressor.fit(X, y)
```
